```python
from matplotlib import pyplot as plt
import math
import numpy as np
import random as rand
```

```python
results = np.array([])
```

```python
def snoepAutomaat(n):
    doosje_1 = n
    doosje_2 = n
    global results
    vol = True
    
    while vol == True:
        keuze = rand.randint(1,2)
        if keuze == 1:
            doosje_1 = doosje_1 - 1
        else: 
            doosje_2 = doosje_2 - 1
        if doosje_1 == -1 or doosje_2 == -1:
            results = np.append(results, doosje_1 + doosje_2 + 1)
            vol = False
        else: continue
```

```python
def testOmgeving(n):
    for i in range(200000):
        snoepAutomaat(n)
    
    global results
    bins = []

    for r in range(48): # aanpassen naargelang n
        bins.append(r/2) # pas aan voor dikte bar
    
    plt.hist(results, bins, align='left', weights=np.ones(len(results)) / len(results))
    plt.title("Verdeling voor n = 30")
    plt.xlabel("k-waarde")
    plt.ylabel("Relatief aantal")
    plt.savefig("verdeling_voor_n_30.svg")
    plt.show
```

```python
testOmgeving(30)
```

Dit volgt een **negatief-binomiale verdeling** met *p = q = 1/2* en *2n - k over n* als binomiaal gedeelte.
